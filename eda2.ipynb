```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
for col in data.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(6, 1.5))
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
data1=data.copy(deep=True)
```

```python
for col in data1.columns:
    q1=data1[col].quantile(0.25)
    q3=data1[col].quantile(0.75)
    iqr=q3-q1
    le=q1-1.5*iqr
    ue=q3+1.5*iqr
    median=data1[col].median()

    data1[col]=np.where((data1[col]<le)|(data1[col]>ue),median,data[col])
```

## statistical Summary

```python
import numpy as np
import pandas as pd

summary=pd.DataFrame()

summary['Mean']=data.mean()
summary['Median']=data.median()
summary['Std Deviation']=data.std()
summary['IQR']=data.quantile(0.75)-data.quantile(0.25)

print(summary)
```
